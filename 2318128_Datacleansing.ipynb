# 3. Normalisasi Nama Toko

```python
def format_nama_toko(df, kolom='Nama_Toko'):
    df[kolom] = df[kolom].str.upper()
    return df

df = format_nama_toko(df)
df.head()
```

# 4. Normalisasi Nama Produk

```python
def format_produk(df, kolom='Produk'):
    df[kolom] = df[kolom].str.title()
    return df

df = format_produk(df)
df.head()
```

# 5. Evaluasi & Standarisasi Nilai Penjualan

```python
def format_total_penjualan(df, kolom='Total_Penjualan'):
    def format_rupiah(x):
        # Ubah ke string dulu
        x = str(x)
        # Bersihkan Rp, titik, koma, dan spasi
        x = x.replace("Rp", "").replace(".", "").replace(",", "").replace(" ", "")
        # Kalau kosong, anggap 0
        if x == "" or x.lower() == "nan":
            return "Rp. 0"
        # Konversi ke int
        nilai = int(x)
        # Format jadi Rp. 1.000.000
        return f"Rp. {nilai:,}".replace(",", ".")

    df[kolom] = df[kolom].apply(format_rupiah)
    return df

# Eksekusi
df = format_total_penjualan(df)
df.head()
```

# 6. Validasi & Perbaikan Nama Kota

```python
def format_nama_kota(df, kolom='Kota'):
    # rapikan spasi + huruf besar di awal kata
    df[kolom] = df[kolom].astype(str).str.strip().str.title()

    # koreksi ejaan khusus
    koreksi = {
        "Makasar": "Makassar"
    }
    df[kolom] = df[kolom].replace(koreksi)

    return df

# Eksekusi
df = format_nama_kota(df, kolom="Kota")
df["Kota"].unique()[:20]  # cek beberapa hasil unik
```

# 7. Standarisasi Kode Negara

```python
def format_negara_kode(df, kolom='Negara_Kode'):
    # rapikan teks dulu
    df[kolom] = df[kolom].astype(str).str.strip().str.title()

    # ubah semua variasi jadi "ID"
    df[kolom] = df[kolom].replace({
        "Indonesia": "ID",
        "Id": "ID",
        "Ind": "ID",
        "Ina": "ID"
    })

    # pastikan semua hasilnya ID
    df[kolom] = "ID"
    return df

# Eksekusi
df = format_negara_kode(df, kolom="Negara_Kode")
df.head()
```

# 8. Enrichment Provinsi dari Kota

```python
def isi_provinsi_dari_kota(df, kolom_kota='Kota', kolom_prov='Provinsi'):
    # Isi provinsi kosong dari nama kot
    kota_to_provinsi = {
    # Sumatera
    "Medan": "Sumatera Utara",
    "Padang": "Sumatera Barat",
    "Palembang": "Sumatera Selatan",
    "Lampung": "Lampung",
    "Batam": "Kepulauan Riau",

    # Jawa
    "Jakarta": "DKI Jakarta",
    "Bekasi": "DKI Jakarta",
    "Depok": "Jawa Barat",
    "Tangerang": "Banten",
    "Bandung": "Jawa Barat",
    "Semarang": "Jawa Tengah",
    "Surakarta": "Jawa Tengah",
    "Yogyakarta": "DI Yogyakarta",
    "Surabaya": "Jawa Timur",

    # Bali & Nusa Tenggara
    "Denpasar": "Bali",

    # Kalimantan
    "Pontianak": "Kalimantan Barat",
    "Banjarmasin": "Kalimantan Selatan",

    # Sulawesi
    "Makassar": "Sulawesi Selatan",
    "Manado": "Sulawesi Utara",

    # Maluku & Papua
    "Ambon": "Maluku",
    "Jayapura": "Papua"
    }
    df[kolom_prov] = df.apply(
        lambda row: kota_to_provinsi.get(row[kolom_kota], row[kolom_prov]),
        axis=1
    )
    return df

# Eksekusi pipeline
df = isi_provinsi_dari_kota(df)   # isi provinsi kosong dari kota
df.head()
```

# 11. Export Data Hasil Cleansing

```python
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
from openpyxl.styles import Alignment, Font, PatternFill, Border, Side

# Path hasil
output_path = "/content/drive/MyDrive/Colab Notebooks/DataPenjualan_Cleansed.xlsx"

# Simpan dulu ke Excel biasa
df.to_excel(output_path, index=False, sheet_name="DataCleansed")

# Buka kembali dengan openpyxl untuk styling
wb = load_workbook(output_path)
ws = wb["DataCleansed"]

# Styling header
header_fill = PatternFill("solid", fgColor="4F81BD")   # warna biru
header_font = Font(color="FFFFFF", bold=True)          # font putih tebal
for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = Alignment(horizontal="center", vertical="center")

# Styling isi (rata tengah)
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, max_col=ws.max_column):
    for cell in row:
        cell.alignment = Alignment(horizontal="center", vertical="center")

# Atur lebar kolom otomatis
for col in ws.columns:
    max_length = 0
    col_letter = get_column_letter(col[0].column)
    for cell in col:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(str(cell.value))
        except:
            pass
    ws.column_dimensions[col_letter].width = max_length + 2

# Tambahkan border tipis
thin_border = Border(
    left=Side(style="thin"),
    right=Side(style="thin"),
    top=Side(style="thin"),
    bottom=Side(style="thin")
)

for row in ws.iter_rows(min_row=1, max_row=ws.max_row, max_col=ws.max_column):
    for cell in row:
        cell.border = thin_border

# Simpan perubahan
wb.save(output_path)

print(f"Hasil cleansing tersimpan dengan tabel rapih di {output_path}")

# Download ke lokal
from google.colab import files
files.download(output_path)
```
